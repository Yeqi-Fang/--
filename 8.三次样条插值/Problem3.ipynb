```python
import numpy as np
from scipy.interpolate import CubicSpline
```

```python
# 三次样条插值

x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 3, 3, 4, 2])
# 构建插值函数
cs = CubicSpline(x, y, bc_type=((2, 0), (2, 0)))
```

```python
from matplotlib import pyplot as plt

x_in = np.linspace(0, 4, 100)
y_in = cs(x_in)
plt.plot(x_in, y_in, label='Interpolation')
plt.plot(x, y, 'o', label='Real data point')
plt.legend()
plt.show()
```
